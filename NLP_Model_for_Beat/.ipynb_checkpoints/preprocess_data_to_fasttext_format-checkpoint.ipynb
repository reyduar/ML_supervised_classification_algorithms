```python
# -*- coding: utf-8 -*-
"""
@author: ariel.duarte

That will read the reviews.json file and write out a text file in fastText format
"""
import pandas as pd
import json
from pathlib import Path

reviews_data = Path("data") / "reviews.json"
fasttext_data = Path("fasttext_dataset.txt")

with reviews_data.open() as input, fasttext_data.open("w") as output:
    for line in input:
        review_data = json.loads(line)

        rating = review_data['stars']
        text = review_data['text'].replace("\n", " ")

        fasttext_line = "__label__{} {}".format(rating, text)

        output.write(fasttext_line + "\n")
```

```python
# load and preview the data
reviews_data = []
with open('data/reviews.json') as f:
    for line in f:
        reviews_data.append(json.loads(line))
rev_df = pd.DataFrame.from_dict(reviews_data)
rev_df.head()
```
